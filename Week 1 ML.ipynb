```python
import pandas as pd
import numpy as np
import matplotlib as mp
import sklearn as sk
import tensorflow as tf
import torch as tr

print("Imported successfully")
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
print("imported Successfully")
```

```python
# Load the Iris Dataset
data = load_iris()
X = data.data #X contains the measurements
y = data.target  #y contains

print(X)
print(y)
```

```python
print(X.shape)
```

```python
#Create a decision Tree Classifier
clf = DecisionTreeClassifier()
print("Done")
```

```python
#Train the model on the entire dataset
clf.fit(X, y)
print("Trained")
```

```python
#Asking the model to predict the species of the first flower in the dataset
prediction = clf.predict([X[0]])                         #The index can be changed when necessary
print(f"Prediction for First sample: {prediction}")
print(f"Actual Species of First sample: {y[0]}")
```

```python
#Asking the model to predict the species of the last flower in the dataset
prediction = clf.predict([X[99]])                         #The index can be changed when necessary
print(f"Prediction for First sample: {prediction}")
print(f"Actual Species of First sample: {y[99]}")
```

```python
from sklearn.model_selection import train_test_split  # split the data into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
print("Trained Successfully")
```

```python
clf.fit(X_train, y_train)   #Train on training set
print("Trained")
```

```python
y_pred = clf.predict(X_test)   #Test on unseen data
print("Tested")
```

```python
from sklearn.metrics import accuracy_score
print("Imported")
```

```python
accuracy = accuracy_score(y_test, y_pred)   #Evaluate Accuracy
print("Final Accuracy: ", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print("Imported")
```

```python
# Display the confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Display detailed classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=data.target_names))
```

```python
# Predict the species of the first flower in the test set
sample_prediction = clf.predict([X_test[0]])
print("\nPrediction for first test sample:", sample_prediction)
print("Actual species of first test sample:", y_test[0])
```
