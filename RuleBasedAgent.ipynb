```python
#List of tuples containing conditions and diagnosis
rules =[("fever and cough", "flu"),
        ("rash and fever", "measles"),
        ("sneezing and runny nose", "cold"),
        ("fever and headache", "malaria"),
]
print("Conditions saved Successfully")
```

```python
def diagnose(symptoms):
    #check rule against provided list of symptoms 
    for condition, diagnosis in rules:
        cond_parts = condition.split(" and ")
        if all (symptom in symptoms for symptom in cond_parts):
            return f"Dignosis: {diagnosis}"

    return "Diagnosis: Unknown"

patient_symptoms = input("Enter your symptoms e.g. fever, cough: ").split(",")
print(diagnose(patient_symptoms))
```
