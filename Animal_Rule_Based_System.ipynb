```python
# List of tuples containing animals and characteristics
animals = [
    ("bark and tail", "dog"),
    ("moo and milk", "Cattle"),
    ("roar and kill", "Wild Cats"),
    ("tall and grass", "Giraffe"),
    ("wings and feathers", "Bird"),
    ("dirty and fat", "Pig"),
    ("wool and calm", "Sheep")
]
print("Animals Saved Successfully")
```

```python
def Identification(features):
    # Normalize user input
    features = [f.strip().lower() for f in features]

    for characteristics, specie in animals:
        charac_parts = characteristics.lower().split(" and ")
        # Check if every expected characteristic is present
        if all(char in features for char in charac_parts):
            return f"Animal: {specie}"
    return "Animal: Unknown"

New_Animal = input('''Enter the Characteristics of the animal
                        e.g. Bark, Wings: ''').split(",")

print(Identification(New_Animal))
```
