```python
import random
print("Imported")
```

```python
# Environment: moods, activities, and playlists
moods = ['Happy', 'Sad', 'Stressed', 'Relaxed']
activities = ['Studying', 'Exercising', 'Commuting', 'Relaxing']
times_of_day = ['Morning', 'Afternoon', 'Evening', 'Night']
print("Environment Saved")
```

```python
playlists = {
    ('Happy', 'Exercising'): ['Upbeat Pop', 'EDM Energy'],
    ('Sad', 'Relaxing'): ['Soft Piano', 'Acoustic Comfort'],
    ('Stressed', 'Studying'): ['Lo-fi Focus', 'Chill Instrumental'],
    ('Relaxed', 'Commuting'): ['Indie Smooth', 'Soft Jazz'],
    ('Happy', 'Commuting'): ['Feel-Good Indie', 'Pop Vibes'],
    ('Relaxed', 'Night'): ['Ambient Calm', 'Sleep Loops'],
    ('Stressed', 'Evening'): ['Chillout Mix', 'Calm Beats']
}
print("Playlist Saved")
```

```python
# Learning agent memory (feedback tracker)
feedback_memory = {}

# Agent decision function
def moodai_agent(mood, activity, time):
    key = (mood, activity)
    if key in playlists:
        options = playlists[key]
    else:
        options = ['Default Mix']

    # Add small adaptive logic
    if key in feedback_memory and feedback_memory[key]['like'] > feedback_memory[key]['dislike']:
        return random.choice(options)
    else:
        # If negative feedback, try random alternative
        return random.choice(list(playlists[random.choice(list(playlists.keys()))]))
```

```python
# Simulation
steps = 6
print("ðŸŽ§ MoodAI Music Recommendation Simulation ðŸŽ§")
print("-" * 60)
```

```python
for step in range(steps):
    mood = random.choice(moods)
    activity = random.choice(activities)
    time = random.choice(times_of_day)

    song = moodai_agent(mood, activity, time)
    print(f"Step {step+1}: Mood={mood}, Activity={activity}, Time={time}")
    print(f">> MoodAI recommends: {song}")

    # Simulate feedback (user likes or dislikes the song)
    feedback = random.choice(['like', 'dislike'])
    print(f"User Feedback: {feedback}")
    key = (mood, activity)
    if key not in feedback_memory:
        feedback_memory[key] = {'like': 0, 'dislike': 0}
    feedback_memory[key][feedback] += 1

    print("Current Feedback Memory:", feedback_memory)
    print("-" * 60)

print("ðŸ§  Final Learning Memory:", feedback_memory)
```
